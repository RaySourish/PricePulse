# ARIMA


ARIMA stands for Autoregressive Integrated Moving Average Model. It belongs to a class of models that explains a given time series based on its own past values -i.e.- its own lags and the lagged forecast errors. The equation can be used to forecast future values. Any ‘non-seasonal’ time series that exhibits patterns and is not a random white noise can be modeled with ARIMA models.


ARIMA Models are specified by three order parameters: (p, d, q)
- p is the order of the AR term

- q is the order of the MA term

- d is the number of differencing required to make the time series stationary

An ARIMA model is one where the time series was differenced at least once to make it stationary and we combine the AR and the MA terms. So the equation of an ARIMA model becomes

## ADF (Augmented Dickey-Fuller) Test
One of the most widely used statistical tests is the Dickey-Fuller test. It can be used to determine whether or not a series has a unit root, and thus whether or not the series is stationary , method to check the null hopethesis using
p-value


```python

def adf_test(time_series):
    result = adfuller(time_series)
    adf_statistic = result[0]
    p_value = result[1]

    print('ADF Statistic:', adf_statistic)
    print('p-value:', p_value)

    if p_value <= 0.05:
        print('The time series is stationary.')
    else:
        print('The time series is non-stationary.')


```

```python
adf_test(data['Close'])
```

```python
def test_stationarity(timeseries):
    #Determing rolling statistics
    rolmean = timeseries.rolling(12).mean()
    rolstd = timeseries.rolling(12).std()
    #Plot rolling statistics:
    plt.plot(timeseries, color='blue',label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean and Standard Deviation')
    plt.show(block=False)
    print("Results of dickey fuller test")
    adft = adfuller(timeseries,autolag='AIC')
    # output for dft will give us without defining what the values are.
    #hence we manually write what values does it explains using a for loop
    output = pd.Series(adft[0:4],index=['Test Statistics','p-value','No. of lags used','Number of observations used'])
    for key,values in adft[4].items():
        output['critical value (%s)'%key] =  values
    print(output)
test_stationarity(data['Close'])
```

```python
import os
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pmdarima as pm
from pylab import rcParams
from statsmodels.tsa.arima_model import ARIMA
from pmdarima.arima import ADFTest
import random
```

```python
def arima(stock_name, data):
    df_close = data['Close']

    # Split data into train and test set (90% - train, 10% - test)
    df_log = df_close
    train_data, test_data = df_log[3:int(len(df_log) * 0.9)], df_log[int(len(df_log) * 0.9):]
    test_values = len(df_log) * 0.01 + 1.0
    tot = len(train_data)+len(test_data )
    x_train = list(range(0, int(len(train_data))))
    x_test = list(range(int(len(train_data)), tot))

    plt.figure(figsize=(16,6))
    plt.plot(x_train, train_data, color='#0077B6', linestyle='-',  label='Training Data')
    plt.plot(x_test, test_data, color='#FF5733', linestyle='-', label='Test Data')
    plt.ylabel('Y-axis')
    plt.title('Plot of Training and Test Data')
    plt.legend()
    plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
    plt.show()

    model = pm.auto_arima(df_log, start_p=0, d=None, start_q=0,
                          max_p=5, max_d=5, max_q=5, start_P=0,
                          D=1, start_Q=0, max_P=5, max_D=5,
                          max_Q=5, m=7, seasonal=True,
                          error_action='warn', trace=True,
                          suppress_warnings=True, stepwise=True,
                          random_state=20, n_fits=50)

    model.summary()

    exo_data = data['Volume']
    exo_data = exo_data[int(len(exo_data) * 0.9):]

    ## prediction for next month Working Days 22
    preds = model.predict(n_periods=22, X=exo_data)
    preds = np.vstack(preds)

    hist_data = yf.download(stock_name, start="2023-06-05", end="2023-07-07")
    hist_data = hist_data.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis=1)
    hist_data = hist_data['Close']
    hist_data = np.array(hist_data)

    rmse = np.sqrt(np.mean(((preds - hist_data) ** 2)))
    print(f'RMSE ARIMA: {rmse}')

    preds_gr = np.reshape(preds, (22,))
    x = list(range(0, 22))
    plt.figure(figsize=(16,6))
    plt.plot(x, hist_data[:22], color='#0077B6', marker='o', label='Historical Data')
    plt.plot(x, preds_gr, color='#FF5733', marker='o', label='Predictions')

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title(f'{stock_name} ARIMA Prediction')
    plt.legend()
    plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
    plt.show()

    return preds, rmse
```

# Stacked LSTM



Long Short-Term Memory (LSTM) is a type of recurrent neural network (RNN) that is designed to overcome the limitations of traditional RNNs in capturing long-term dependencies in sequential data. Unlike standard RNNs, LSTM has a unique memory cell that can store and update information over time, allowing it to retain important information for longer durations. This capability makes LSTM particularly effective in tasks such as time series analysis, natural language processing, and speech recognition, where long-term dependencies play a crucial role.
**Stacked LSTM** refers to the stacking of multiple LSTM (Long Short-Term Memory) layers in a neural network architecture.It helps in capturing complex temporal patterns. By stacking LSTM layers, the model can learn hierarchical representations of the input data, enabling it to capture intricate relationships at different levels of abstraction

```python
##splitting dataset into train and test split
training_size=int(len(df1)*0.80)
test_size=len(df1)-training_size
train_data,test_data=df1[0:training_size,:],df1[training_size:len(df1),:1]
```

```python
training_size,test_size
```

```python
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, ytest = create_dataset(test_data, time_step)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
print(X_test.shape), print(ytest.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
### Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,ytest),epochs=100,batch_size=64,verbose=1)
```

## Prediction

```python
###  prediction and check performance metrics
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
##Transformback to original form
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
### RMSE performance metrics
import math
from sklearn.metrics import mean_squared_error
print("Rmse for train_data :" )
print (math.sqrt(mean_squared_error(y_train,train_predict)))
print("Rmse for test_data :" )
math.sqrt((mean_squared_error(ytest,test_predict)))
```

- ### The **LSTM and ARIMA** models exhibit distinct architectural features and provide differing predictions for the market. The LSTM model, being a type of recurrent neural network (RNN), is capable of capturing complex nonlinear patterns and long-term dependencies in time series data. Its architecture includes memory cells and gates that can retain and selectively update information over time. However, in this particular case, the LSTM model is providing predictions indicating a market fall, which may be attributed to the patterns it has learned from the training data.

- ### On the other hand, the **ARIMA model** is a traditional time series model that analyzes autocorrelation and seasonality patterns in the data. It uses a combination of autoregressive (AR), differencing (I), and moving average (MA) components to capture the underlying patterns. The ARIMA model's prediction aligns more closely with the actual market behavior, indicating its ability to capture the linear relationships and stationary properties of the time series.
